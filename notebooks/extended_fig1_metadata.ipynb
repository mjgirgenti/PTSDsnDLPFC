# Extended Data Figure 1a: 111 RNA and ATAC metadata table

```python
meta = pd.read_csv('/gpfs/gibbs/pi/gerstein/jz435/ShareZhangLab/PTSD/supp_tables/csvs/Supp_Table1_Sample_metadata.csv',sep='\t')
```

```python
meta = meta[meta.Modality=='RNA/ATAC']
```

```python
meta
```

```python
con = meta[meta.Condition=='CON']
mdd = meta[meta.Condition=='MDD']
ptsd = meta[meta.Condition=='PTSD']
```

```python
sex_all = []
race_all = []
age_all = []
pmi_all = []
rin_all = []
smoking_all = []
anti_all = []

for data in [con,mdd,ptsd]:
    sex = round(data[data.Sex=='M'].shape[0]/data.shape[0],2)
    race = round(data[data.Race=='W'].shape[0]/data.shape[0],2)
    age = str(round(data.AgeDeath.describe()['mean'],2)) + ' Â± ' + str(round(con['AgeDeath'].describe()['std'],2))
    pmi = str(round(data.PMI.describe()['mean'],2)) + ' Â± ' + str(round(con['PMI'].describe()['std'],2))
    rin = str(round(data.RIN.describe()['mean'],2)) + ' Â± ' + str(round(con['RIN'].describe()['std'],2))
    smoking = round(data[data.Smoking=='Y'].shape[0]/data.shape[0],2)
    anti = round(data[data.Antidepressant=='Y'].shape[0]/data.shape[0],2)
    sex_all.append(sex)
    race_all.append(race)
    age_all.append(age)
    pmi_all.append(pmi)
    rin_all.append(rin)
    smoking_all.append(smoking)
    anti_all.append(anti)
```

```python
df = pd.DataFrame(data=[sex_all,race_all,age_all,pmi_all,rin_all,smoking_all,anti_all],
             columns=['CON (N=39)','MDD (N=36)','PTSD (N=36)'],
             index=['Sex (% Male)','Race (% Caucasian)','Age at Death (mean Â± std)','PMI (mean Â± std)','RIN (mean Â± std)','Smoking (% Yes)','Antidepressant (% Yes)'])
```

```python
pdf_table = []
pdf_table.append(df)
```

```python
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42
plt.rcParams['font.size'] = '16'
fig, ax =plt.subplots(figsize=(7,7))
ax.axis('tight')
ax.axis('off')
the_table = ax.table(cellText=df.values,colLabels=df.columns,rowLabels=df.index,loc='center')
the_table.scale(1,2)
```
