# Save peaks as bed files

```python
clusters <- c('Ex','In','Oli','OPC','End','Ast','Mic')
for (cluster in clusters) {
    peaks <- readRDS(glue('/gpfs/gibbs/pi/gerstein/jz435/ShareZhangLab/PTSD/ATAC/data/ATAC_FINAL/PeakCalls/{cluster}-reproduciblePeaks.gr.rds'))
    peaks_merged <- unlist(as(peaks, "GRangesList"))
    peaks_merged_sorted <- sort(peaks_merged)
    peaks_merged_sorted_df <- data.frame(peaks_merged_sorted)
    print(dim(peaks_merged_sorted_df))
    fwrite(peaks_merged_sorted_df,file=glue('/gpfs/gibbs/pi/gerstein/jz435/ShareZhangLab/PTSD/ATAC/data/ATAC_FINAL/PeakBeds/{cluster}.peaks.bed'), quote=F, sep="\t", row.names=F, col.names=F)
}
```

# Extended Data Figure 6a: ATAC marker peaks

```python
markersPeaks <- loadHDF5SummarizedExperiment('/gpfs/gibbs/pi/gerstein/jz435/ShareZhangLab/PTSD/ATAC/data/ATAC_FINAL/MarkersPeaks')
```

```python
markerList <- getMarkers(markersPeaks, cutOff = "FDR <= 0.05 & Log2FC >= 1")
```

```python
markersPeaks <- getMarkerFeatures(
    ArchRProj = proj, 
    useMatrix = "PeakMatrix", 
    groupBy = "Int_Cluster",
  bias = c("TSSEnrichment", "log10(nFrags)"),
  testMethod = "wilcoxon"
)
```

```python
heatmapPeaks <- plotMarkerHeatmap(
  seMarker = markersPeaks, 
  cutOff = "FDR <= 0.01 & Log2FC >= 0.7",
  transpose = TRUE
)
```

```python
draw(heatmapPeaks, heatmap_legend_side = "bot", annotation_legend_side = "bot")
```

```python
plotPDF(heatmapPeaks, name = "Peak-Marker-Heatmap", width = 8, height = 6, ArchRProj = proj, addDOC = FALSE)
```

# Figure 1a: P2G

```python
genes_df = data.frame(getGenes(proj))
genes_df = na.omit(genes_df[genes_df$symbol=='CAMKV',])

p2g <- getPeak2GeneLinks(proj,resolution = NULL, corCutOff = 0.6, FDRCutOff=0.1, returnLoops = T)
df <- data.frame(p2g$Peak2GeneLinks)
loops <- df[df$seqnames==df$seqnames & df$start > df$start - 50000 & df$end < df$end + 50000,]

loops <- loops[,c('seqnames','start','end','value')]
dir <- genes_df$strand
if (dir=='-') {
    loops <- rbind(loops[loops$start==genes_df$end,],loops[loops$end==genes_df$end,])
} else {
    loops <- rbind(loops[loops$start==genes_df$start,],loops[loops$end==genes_df$start,])
}
loops
```

```python
p <- plotBrowserTrackFKBP5(
    ArchRProj=proj,
    groupBy='Int_Cluster',
    geneSymbol='CAMKV',
    #region = GRanges(seqnames=genes_df$seqnames,ranges=IRanges(start=df$end - 100000,end=df$end + 100000)),
    loops = GRanges(loops),
    plotSummary = c('featureTrack','loopTrack','geneTrack'),
    sizes = c(0.1, 0.2, 0.2),
    borderWidth=1
)
```

```python
plotPDF(p, 
    name = "Fig1a_P2G.pdf", 
    ArchRProj = proj, 
    addDOC = FALSE, width = 4, height = 2)
```
