```python
def plot_colortable(colors, *, ncols=4, sort_colors=True):
    # Sort colors by hue, saturation, value and name.
    if sort_colors is True:
        names = sorted(
            colors, key=lambda c: tuple(mcolors.rgb_to_hsv(mcolors.to_rgb(c))))
    else:
        names = list(colors)
    return names
```

```python
def create_metadata_df(name):
    arr = []
    for s in meta.subclass_renamed.cat.categories:
        print(s)
        vals = list(meta[meta.subclass_renamed==s][name].value_counts()[sorted(meta[name].unique())]/meta[meta.subclass_renamed==s][name].value_counts()[sorted(meta[name].unique())].sum())
        arr.append(vals)
    df = pd.DataFrame(arr)
    df.insert(0, "subclass", meta.subclass_renamed.cat.categories, True)
    df.columns = ['subclass']+sorted(meta[name].unique())
    return df 
```

```python
def metadata_barplot(df,savefile,cmap):
    plt.rcParams['pdf.fonttype'] = 42
    plt.rcParams['ps.fonttype'] = 42
    #df.plot('subclass',kind='bar',stacked=True,rot=45,fontsize=16,figsize=(7,5),legend=False)
    if cmap is not None:
        df.plot('subclass',kind='bar',stacked=True,rot=45,fontsize=16,figsize=(7,5),legend=False,cmap=cmap)
    else:
        df.plot('subclass',kind='bar',stacked=True,rot=45,fontsize=16,figsize=(7,5),legend=False)
    plt.xticks(np.arange(len(meta.subclass_renamed.cat.categories)), meta.subclass_renamed.cat.categories, ha='right',rotation_mode='anchor')
    plt.ylabel('Proportion',fontsize=16)
    plt.xlabel(' ')
    plt.savefig(savefile,bbox_inches='tight',dpi=300)
```

```python
def create_metadata_df_bins(name):
    arr = []
    for s in meta.subclass_renamed.cat.categories:
        print(s)
        temp = pd.cut(meta[meta.subclass_renamed==s][name],bins=10)
        vals = list(temp.value_counts()[sorted(temp.cat.categories)]/meta[meta.subclass_renamed==s][name].value_counts().sum())
        arr.append(vals)
    df = pd.DataFrame(arr)
    df.insert(0, "subclass", meta.subclass_renamed.cat.categories, True)
    df.columns = ['subclass']+sorted(temp.cat.categories)
    return df 
```

```python
df = create_metadata_df_bins('RIN')
metadata_barplot(df,'/home/ah2428/supp/fig1/rin_barplot.pdf','Purples')
```

```python
df = create_metadata_df_bins('PMI')
metadata_barplot(df,'/home/ah2428/supp/fig1/pmi_barplot.pdf','Greens')
```
