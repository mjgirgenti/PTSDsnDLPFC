```python
name = 'Sample'
arr = []
for s in meta.subclass.cat.categories:
    print(s)
    vals = list(meta[meta.subclass==s][name].value_counts()[sorted(meta[name].unique())]/meta[meta.subclass==s][name].value_counts()[sorted(meta[name].unique())].sum())
    arr.append(vals)
df = pd.DataFrame(arr)
df.insert(0, "subclass", meta.subclass.cat.categories, True)
df.columns = ['subclass']+sorted(meta[name].unique())
```

```python
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42
df.plot('subclass',kind='bar',stacked=True,rot=45,fontsize=16,figsize=(7,5),legend=False,color=colors)
plt.xticks(np.arange(len(meta.subclass.cat.categories)), meta.subclass.cat.categories, ha='right',rotation_mode='anchor')
plt.ylabel('Proportion',fontsize=16)
plt.xlabel(' ')
plt.savefig('/home/ah2428/supp/multiome/sample_barplot.pdf',bbox_inches='tight',dpi=300)
```
