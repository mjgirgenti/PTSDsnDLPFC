# Extended Data Figure 4a: PTSD vs MDD DEGs venn

```python
ptsd_degs = []
for s in ['CUX2','RORB','FEZF2','OPRK1','LAMP5','KCNG1','VIP','SST','PVALB','OLI','OPC','END','AST','MIC']:
    mast = pd.read_csv(f'/gpfs/gibbs/pi/girgenti/JZhang/commonData/PTSD/RNA/results/062522_DEG/subclass/mast/PTSD_vs_CON/{s}_SIG_DEG.csv',sep='\t')
    
    mast = mast[np.abs(mast.coef) > np.log2(1.2)]
    mast_up = mast[mast.coef > 0].primerid.values
    mast_down = mast[mast.coef < 0].primerid.values
    
    wilcox = pd.read_csv(f'/gpfs/gibbs/pi/girgenti/JZhang/commonData/PTSD/RNA/results/062522_DEG/subclass/wilcox/PTSD_vs_CON/{s}_SIG_DEG.csv',sep='\t')
    wilcox = wilcox[np.abs(wilcox.avg_log2FC) > np.log2(1.2)]
    wilcox_up = wilcox[wilcox.avg_log2FC > 0].index.values
    wilcox_down = wilcox[wilcox.avg_log2FC < 0].index.values
    
    inter_up = list(set(mast_up) & set(wilcox_up))
    inter_down = list(set(mast_down) & set(wilcox_down))
    inter_up = [x for x in inter_up if x in protein_coding]
    inter_down = [x for x in inter_down if x in protein_coding]

    degs = np.append(inter_up,inter_down)
    ptsd_degs.append(degs)
```

```python
mdd_degs = []
for s in ['CUX2','RORB','FEZF2','OPRK1','LAMP5','KCNG1','VIP','SST','PVALB','OLI','OPC','END','AST','MIC']:
    mast = pd.read_csv(f'/gpfs/gibbs/pi/girgenti/JZhang/commonData/PTSD/RNA/results/062522_DEG/subclass/mast/MDD_vs_CON/{s}_SIG_DEG.csv',sep='\t')
    
    mast = mast[np.abs(mast.coef) > np.log2(1.2)]
    mast_up = mast[mast.coef > 0].primerid.values
    mast_down = mast[mast.coef < 0].primerid.values
    
    wilcox = pd.read_csv(f'/gpfs/gibbs/pi/girgenti/JZhang/commonData/PTSD/RNA/results/062522_DEG/subclass/wilcox/MDD_vs_CON/{s}_SIG_DEG.csv',sep='\t')
    wilcox = wilcox[np.abs(wilcox.avg_log2FC) > np.log2(1.2)]
    wilcox_up = wilcox[wilcox.avg_log2FC > 0].index.values
    wilcox_down = wilcox[wilcox.avg_log2FC < 0].index.values
    
    inter_up = list(set(mast_up) & set(wilcox_up))
    inter_down = list(set(mast_down) & set(wilcox_down))
    inter_up = [x for x in inter_up if x in protein_coding]
    inter_down = [x for x in inter_down if x in protein_coding]

    degs = np.append(inter_up,inter_down)
    mdd_degs.append(degs)
```

```python
exc_ptsd_degs = list(set(list(itertools.chain(*ptsd_degs[:4]))))
exc_mdd_degs = list(set(list(itertools.chain(*mdd_degs[:4]))))
```

```python
inh_ptsd_degs = list(set(list(itertools.chain(*ptsd_degs[4:9]))))
inh_mdd_degs = list(set(list(itertools.chain(*mdd_degs[4:9]))))
```

```python
ptsds = [exc_ptsd_degs,inh_ptsd_degs,ptsd_degs[9],ptsd_degs[10],ptsd_degs[11],ptsd_degs[12],ptsd_degs[13]]
mdds = [exc_mdd_degs,inh_mdd_degs,mdd_degs[9],mdd_degs[10],mdd_degs[11],mdd_degs[12],mdd_degs[13]]
```

```python
for p,m,c in zip(ptsds,mdds,celltypes):
    inter = len(list(set(p) & set(m)))
    plt.figure()
    venn2(subsets=(len(p)-inter, len(m)-inter, inter), set_labels=(f'PTSD {c}',f'MDD {c}'),set_colors=('tab:blue','tab:orange'),alpha=0.6)
    venn2_circles(subsets=(len(p)-inter, len(m)-inter, inter),linewidth=0.5)
```

# Extended Data Figure 4b: PTSD UP MDD DOWN

```python
# SIG DEG for PTSD, DEG for MDD

ptsd_ups = []
ptsd_downs = []
mdd_ups = []
mdd_downs = []
for s in subclasses_old:
    print(s)
    ptsd_mast = pd.read_csv(f'/gpfs/gibbs/pi/girgenti/JZhang/commonData/PTSD/RNA/results/062522_DEG/subclass/mast/PTSD_vs_CON/{s}_SIG_DEG.csv',sep='\t')
    ptsd_wilcox = pd.read_csv(f'/gpfs/gibbs/pi/girgenti/JZhang/commonData/PTSD/RNA/results/062522_DEG/subclass/wilcox/PTSD_vs_CON/{s}_SIG_DEG.csv',sep='\t')
    
    ptsd_mast_up = ptsd_mast[ptsd_mast.coef > 0]
    ptsd_wilcox_up = ptsd_wilcox[ptsd_wilcox.avg_log2FC > 0]
    inter = list(set(ptsd_mast_up.primerid)&set(ptsd_wilcox_up.index))
    x_up = ptsd_mast_up[ptsd_mast_up.primerid.isin(inter)]
    ptsd_up_celltype = pd.DataFrame({'condition':'PTSD','subclass':s,'DEG':x_up.primerid.values,'fc':x_up.coef.values,'FDR':x_up.fdr.values}).reset_index(drop=True)
    print(ptsd_up_celltype.shape)
    
    ptsd_mast_down = ptsd_mast[ptsd_mast.coef < 0]
    ptsd_wilcox_down = ptsd_wilcox[ptsd_wilcox.avg_log2FC < 0]
    inter = list(set(ptsd_mast_down.primerid)&set(ptsd_wilcox_down.index))
    x_down = ptsd_mast_down[ptsd_mast_down.primerid.isin(inter)]
    ptsd_down_celltype = pd.DataFrame({'condition':'PTSD','subclass':s,'DEG':x_down.primerid.values,'fc':x_down.coef.values,'FDR':x_down.fdr.values}).reset_index(drop=True)
    
    mdd_mast = pd.read_csv(f'/gpfs/gibbs/pi/girgenti/JZhang/commonData/PTSD/RNA/results/062522_DEG/subclass/mast/MDD_vs_CON/{s}_DEG.csv',sep='\t')
    mdd_wilcox = pd.read_csv(f'/gpfs/gibbs/pi/girgenti/JZhang/commonData/PTSD/RNA/results/062522_DEG/subclass/wilcox/MDD_vs_CON/{s}_DEG.csv',sep='\t')
    
    mdd_mast_up = mdd_mast[mdd_mast.coef > 0]
    mdd_wilcox_up = mdd_wilcox[mdd_wilcox.avg_log2FC > 0]
    inter = list(set(mdd_mast_up.primerid)&set(mdd_wilcox_up.index))
    x_up = mdd_mast_up[mdd_mast_up.primerid.isin(inter)]
    mdd_up_celltype = pd.DataFrame({'condition':'MDD','subclass':s,'DEG':x_up.primerid.values,'fc':x_up.coef.values,'FDR':x_up.fdr.values}).reset_index(drop=True)
    
    mdd_mast_down = mdd_mast[mdd_mast.coef < 0]
    mdd_wilcox_down = mdd_wilcox[mdd_wilcox.avg_log2FC < 0]
    inter = list(set(mdd_mast_down.primerid)&set(mdd_wilcox_down.index))
    x_down = mdd_mast_down[mdd_mast_down.primerid.isin(inter)]
    mdd_down_celltype = pd.DataFrame({'condition':'MDD','subclass':s,'DEG':x_down.primerid.values,'fc':x_down.coef.values,'FDR':x_down.fdr.values}).reset_index(drop=True)
    
    ptsd_ups.append(ptsd_up_celltype)
    ptsd_downs.append(ptsd_down_celltype)
    mdd_ups.append(mdd_up_celltype)
    mdd_downs.append(mdd_down_celltype)
```
