```python
name = 'Sample'
arr = []
for s in meta.Int_Cluster.cat.categories:
    print(s)
    vals = list(meta[meta.Int_Cluster==s][name].value_counts()[sorted(meta[name].unique())]/meta[meta.Int_Cluster==s][name].value_counts()[sorted(meta[name].unique())].sum())
    arr.append(vals)
df = pd.DataFrame(arr)
df.insert(0, "subclass", meta.Int_Cluster.cat.categories, True)
df.columns = ['subclass']+sorted(meta[name].unique())
```

```python
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42
df.plot('subclass',kind='bar',stacked=True,rot=45,fontsize=16,figsize=(7,5),legend=False,color=colors)
plt.ylabel('Proportion',fontsize=16)
plt.xlabel(' ')
plt.savefig('/home/ah2428/supp/atac/sample_barplot.pdf',bbox_inches='tight',dpi=300)
```
