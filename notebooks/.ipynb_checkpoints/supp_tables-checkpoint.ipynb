```python
genes = list(set(np.append(mast.Genename,wilcox.Genename)))
```

```python
final_dfs = []
for g in genes:
    mast_g = mast[mast.Genename==g]
    wilcox_g = wilcox[wilcox.Genename==g]
    
    mast_fcs = []
    mast_fdrs = []
    for s in subclasses_old:
        if s in mast_g.Subclass.unique():
            mast_fc = mast_g[mast_g.Subclass==s].log2FC.values[0]
            mast_fdr = mast_g[mast_g.Subclass==s].FDR.values[0]
        else:
            mast_fc = 'not sig'
            mast_fdr = 'not sig'
        mast_fcs.append(mast_fc)
        mast_fdrs.append(mast_fdr)
        
    wilcox_fcs = []
    wilcox_fdrs = []
    for s in subclasses_old:
        if s in wilcox_g.Subclass.unique():
            wilcox_fc = wilcox_g[wilcox_g.Subclass==s].log2FC.values[0]
            wilcox_fdr = wilcox_g[wilcox_g.Subclass==s].FDR.values[0]
        else:
            wilcox_fc = 'not sig'
            wilcox_fdr = 'not sig'
        wilcox_fcs.append(wilcox_fc)
        wilcox_fdrs.append(wilcox_fdr)
        
    mast_df = pd.DataFrame(data={'Genename':[g]*len(subclasses_old),'Subclass':subclasses_old,'MAST log2FC':mast_fcs,'MAST FDR':mast_fdrs})
    wilcox_df = pd.DataFrame(data={'Wilcox log2FC':wilcox_fcs,'Wilcox FDR':wilcox_fdrs})
    final_df = pd.concat([mast_df,wilcox_df],axis=1)
    final_df = final_df[~((final_df['MAST log2FC']=='not sig') & (final_df['Wilcox log2FC']=='not sig'))]
    final_df = final_df.reset_index(drop=True)
    final_dfs.append(final_df)
```

```python
def create_GO_df(file,analysis_name):
    data = pd.read_csv(file,sep='\t')
    name = f.split('/')[-1].split('.')[0].split('_')[1]
    if name=='BP':
        ontology = 'GO_Biological_Process_2023'
    elif name=='CC':
        ontology = 'GO_Cellular_Component_2023'
    elif name=='MF':
        ontology = 'GO_Molecular_Function_2023'
    elif name=='SynGO':
        ontology = 'SynGO 2022'
    else:
        pass
    data['Ontology'] = ontology
    data['-log10FDR'] = -np.log10(data['Adjusted P-value'])
    data['Analysis'] = analysis_name
    data = data.sort_values('-log10FDR',ascending=False)
    return data
```

```python
dfs = []
for f in glob.glob('/gpfs/gibbs/pi/gerstein/jz435/ShareZhangLab/PTSD/RNA/GO/1142_GO/*'):
    df = create_GO_df(f,'1142 PTSD unique DEGs')
    dfs.append(df)

for celltype in ['EXC','INH','OLI','OPC','END','AST','MIC']:
    for f in glob.glob(f'/gpfs/gibbs/pi/gerstein/jz435/ShareZhangLab/PTSD/RNA/GO/celltype_GO/enrichr/{celltype}_*'):
        df = create_GO_df(f,f'{celltype} cell type-specific DEGs')
        dfs.append(df)

for f in glob.glob('/gpfs/gibbs/pi/gerstein/jz435/ShareZhangLab/PTSD/RNA/GO/PTSD_299_GO/*'):
    df = create_GO_df(f,'299 PTSD only DEGs')
    dfs.append(df)
        
for f in glob.glob('/gpfs/gibbs/pi/gerstein/jz435/ShareZhangLab/PTSD/RNA/GO/PTSD_vs_MDD_div_GO/*'):
    df = create_GO_df(f,'104 PTSD vs MDD divergent DEGs')
    dfs.append(df)
```

```python
data
```

```python
dfs = []
for celltype in celltypes:
    d = data[data.Celltype==celltype]
    tfs = []
    ndegs = []
    for tf in d.TF.unique():
        tf_df = d[d.TF==tf]
        tf_df = tf_df[tf_df.DEG.isin(['UP','DOWN'])]
        ndeg = len(tf_df.geneName.unique())
        tfs.append(tf)
        ndegs.append(ndeg)
    df = pd.DataFrame(tfs,ndegs).reset_index()
    df.columns = ['Number of DEGs',f'{celltype} TF']
    df = df[[f'{celltype} TF','Number of DEGs']]
    df = df.sort_values('Number of DEGs',ascending=False).reset_index(drop=True)
    dfs.append(df)
dfs = pd.concat(dfs,axis=1)
```
