# Extended Data Figure 3a: Cosine similarity between 14 DEG lists

```python
def counter_cosine_similarity(c1, c2):
    terms = set(c1).union(c2)
    dotprod = sum(c1.get(k, 0) * c2.get(k, 0) for k in terms)
    magA = math.sqrt(sum(c1.get(k, 0)**2 for k in terms))
    magB = math.sqrt(sum(c2.get(k, 0)**2 for k in terms))
    return dotprod / (magA * magB)
```

```python
degs = []
for i in subclasses_old:
    up = pd.read_csv(f'/gpfs/gibbs/pi/gerstein/jz435/ShareZhangLab/PTSD/RNA/mast_wilcox_intersected_DEG/{i}_PTSD_UP.csv',header=None)[0].values
    down = pd.read_csv(f'/gpfs/gibbs/pi/gerstein/jz435/ShareZhangLab/PTSD/RNA/mast_wilcox_intersected_DEG/{i}_PTSD_DOWN.csv',header=None)[0].values
    deg = np.append(up,down)
    degs.append(deg)
```

```python
sim = []
for i in degs:
    for j in degs:
        s = counter_cosine_similarity(Counter(i),Counter(j))
        sim.append(s)
```

```python
sim = np.array(sim).reshape(14,14)
```

```python
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42

fig = plt.figure(figsize=(8,8))

ax1 = fig.add_axes([0.1,0.1,0.8,0.6])

celltypes = subclasses

plt.rcParams['font.size']= '16'

ax1.axis('off')
ax1.set_xticks([])
ax1.set_yticks([])

axmatrix = fig.add_axes([0.3,0.1,0.6,0.6])
im = axmatrix.matshow(sim, aspect='auto', cmap='Reds')
axmatrix.set(xticks=np.arange(len(celltypes)),xticklabels=celltypes)
axmatrix.set_xticklabels(celltypes,rotation=90)
axmatrix.tick_params(axis="x", labelbottom=True, labeltop=False)
axmatrix.set(yticks=np.arange(len(celltypes)),yticklabels=celltypes)
axmatrix.tick_params(axis=u'both', which=u'both',length=0)

axcolor = fig.add_axes([0.92,0.1,0.03,0.6])
cbar = plt.colorbar(im, cax=axcolor)
cbar.set_ticks([sim.min(),sim.max()])
cbar.set_ticklabels([np.round(sim.min(),2),np.round(sim.max(),2)])

for i in range(14):
    for j in range(14):
        axmatrix.text(i,j,np.round(sim[i,j],2),ha="center", va="center",fontsize=10)
        
plt.savefig('/home/ah2428/deg_cosine_sim.pdf',bbox_inches='tight',dpi=300)
```

```python
all_genes = []

terms = ['Neuron Projection (GO:0043005)',
                 'Regulation Of Cell Migration (GO:0030334)',
                 'Modulation Of Chemical Synaptic Transmission (GO:0050804)',
                 'Axon Guidance (GO:0007411)',
                 'Cell-Cell Junction (GO:0005911)',
                 'Proton-Transporting V-type ATPase, V1 Domain (GO:0033180)',
                 'Nervous System Development (GO:0007399)',
                 'Metal Ion Binding (GO:0046872)',
                 'Kinase Binding (GO:0019900)',
                 'Memory (GO:0007613)',
                 'Dendrite (GO:0030425)',
                 'Calcium Ion Binding (GO:0005509)',
                 'Cytoplasmic Vesicle Membrane (GO:0030659)',
                 'Polymeric Cytoskeletal Fiber (GO:0099513)',
                 'Regulation Of Focal Adhesion Assembly (GO:0051893)']
for i in terms:
    genes = df.loc[i]['Genes'].split(';')
    all_genes.append(genes)
```

```python
all_genes = list(itertools.chain(*all_genes))
```

```python
df_genes = pd.Series(all_genes).value_counts()[:80].index
```

```python
mat = np.zeros((len(terms),len(df_genes)))
for ii,term in enumerate(terms):
    for jj,gene in enumerate(df_genes):
        go_genes = df.loc[term].Genes.split(';')
        if gene in go_genes:
            mat[ii,jj] = df.loc[term]['-log10FDR']
        else:
            pass
```

```python
mat_df = pd.DataFrame(mat,index=renamed,columns=df_genes)
mat_df = mat_df.replace(0, np.nan)
mat_df = mat_df.dropna(axis=1, how='all')
```
